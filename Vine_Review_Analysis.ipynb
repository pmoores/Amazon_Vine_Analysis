```python
# Module 16 - Big Data (Deliverable 2 - Pandas Analysis)
# Import Pandas dependency
import pandas as pd
```

```python
# Load Vine CSV file
vine_data_to_load = "Resources/Vine_table.csv"
```

```python
# Read in the Vine csv file as a DataFrame, remove index, then preview
vine_df = pd.read_csv(vine_data_to_load, index_col=False)
vine_df.set_index('review_id', inplace=True)
vine_df.head()
```

```python
# 1. Create a DataFrame where total_votes is greater than or equal to 20
total_votes_filtered_df = vine_df.loc[vine_df['total_votes'] >= 20]
total_votes_filtered_df.sample(n=7)
```

```python
# 2. Create a DataFrame from total_votes_filtered_df where helpful_votes/total_votes is >= 50%
helpful_by_total_df = total_votes_filtered_df[total_votes_filtered_df['helpful_votes']/total_votes_filtered_df['total_votes'] >= 0.5]
helpful_by_total_df.sample(n=7)
```

```python
# 3. Create a DataFrame from helpful_by_total_df where a review was part of Vine
vine_yes_df = helpful_by_total_df.loc[helpful_by_total_df['vine'] == 'Y']
vine_yes_df.sample(n=7)
```

```python
# 4. Create a DataFrame from helpful_by_total_df where a review was NOT part of Vine
vine_no_df = helpful_by_total_df.loc[helpful_by_total_df['vine'] == 'N']
vine_no_df.sample(n=7)
```

```python
# Total number of Vine reviews
total_vine_reviews = vine_yes_df['vine'].count()
print(total_vine_reviews)
```

```python
# Total number of non-Vine reviews
total_nonvine_reviews = vine_no_df['vine'].count()
print(total_nonvine_reviews)
```

```python
# Number of 5-star Vine reviews
vine_5star_df = vine_yes_df.loc[vine_yes_df['star_rating'] == 5.0]
total_vine_5star = vine_5star_df['star_rating'].count()
print(total_vine_5star)
```

```python
# Number of 5-star non-Vine reviews
# Number of 5-star Vine reviews
nonvine_5star_df = vine_no_df.loc[vine_no_df['star_rating'] == 5.0]
total_nonvine_5star = nonvine_5star_df['star_rating'].count()
print(total_nonvine_5star)
```

```python
# Percentage of 5-star Vine reviews
percent_5star_vine = (total_vine_5star/total_vine_reviews)*100
print(percent_5star_vine)
```

```python
# Percentage of 5-star non-Vine reviews
percent_5star_nonvine = (total_nonvine_5star/total_nonvine_reviews)*100
print(percent_5star_nonvine)
```

```python
# Print summary statement
Summary_statement = (f'Total Number of Vine Reviews: {total_vine_reviews}\n'
                     f'Total Number of Non-Vine Reviews: {total_nonvine_reviews}\n'
                     f'Total Number of Vine 5-Star Reviews: {total_vine_5star}\n'
                     f'Total Number of Non-Vine 5-Star Reviews: {total_nonvine_5star}\n'
                     f'Percentage of 5-Star Vine Reviews: {percent_5star_vine:.2f}\n'
                     f'Percentage of 5-Star Non-Vine Reviews: {percent_5star_nonvine:.2f}\n')

print(Summary_statement)
```
